

# Anomaly Detection using Gaussian Probabilty

```
import numpy as np
import matplotlib.pyplot as plt
```

```
mu = 0
sigma = 1
x1 = np.random.normal(mu, sigma, size=1000)
x2 = np.random.normal(mu, sigma, size=1000)
```

```
plt.figure(figsize=(10,10))
plt.scatter(x1, x2)
```

```
plt.figure(figsize=(10,10))
n, bins, patches = plt.hist(x=x1, bins=10)
```

```
def gaussian(x, mu=0, sigma=1):
  px = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp( - (x - mu)**2 / (2 * sigma**2) )
  #print(px)
  return px

px1 = []
for i in range(0,1000):
    px1.append(gaussian(x1[i]))

print(px1)
```

```
px2 = []
for i in range(0,1000):
    px2.append(gaussian(x2[i]))

print(px2[0])
```

```
# (1/(sigma * np.sqrt(2 * np.pi))) * np.exp( - (bins - mu)**2 / (2 * sigma**2) )


px = []

for i in range(0,1000):
  temp = px1[i]*px2[i]
  #print(temp)
  px.append(temp)
    #px1.append(px1[i]*px2[i])
    
print(px)
```

```
ep = min(px)
print(ep)
```

prediction time

```
px15 = gaussian(1)
px25 = gaussian(1)
pred = px15*px25
if pred<ep:
  print("Anomaly")
else:
  print("Normal")
```
